```python
from collections import deque
import random
import math
import pandas as pd
import time
import csv
from scipy.spatial import distance
```

```python
csv_file = "../Project#1/2023_AI_TSP.csv"
solution_file = "solution.csv"
GA_Interation = 10000
running_time = 1000 #1조
start_time = time.time()
```

```python
#데이터 좌표를 리스트로 변경
data = pd.read_csv(csv_file, header=None, names=['x', 'y'])

list_data = data.values.tolist() #좌표 리스트

len_data = len(list_data) #행의 개수

print(list_data)
print(len_data)
```

```python
# 거리 계산 함수
# 거리 : node 1과 node2의 직선거리(피타고라스 정리)
# node = list_data의 idex
def distance_between_node(node1, node2):
    distance_x = list_data[node1][0] - list_data[node2][0]
    distance_y = list_data[node1][1] - list_data[node2][1]

    distace_pow = math.pow(distance_x, 2) + math.pow(distance_y, 2)

    return math.sqrt(distace_pow)

print(list_data[0]+list_data[1])
distance_between_node(0, 1)
```

```python
# 경로 평가 함수 -> 모든 노드 순회하며 거리 계산
def evaluate_path(path):
    dist = 0
    for i in range(0, len(path)):
        node1 = path[i]
        node2 = path[(i+1)%len(path)]

        dist += distance_between_node(node1, node2)
    return dist

evaluate_path([0, 1, 2])
```

```python
# 무작위로 해를 만듦
def make_path_list( ):
    path_list = []
    for i in range(running_time):
        #print('path running_time : ', i)
        path = list(range(1, len_data)) #모든 노드의 index를 리스트로 생성
        random.shuffle(path) #노드 index 순서를 섞음
        if path not in path_list:
            path.insert(0, 0)
            path_list.append(path)
    return path_list
```

```python
# 모든 경로를 순회하여 최적해 찾기
def find_best_path(run_count):
    best_path = []
    dist = float('inf')
    for i in range(run_count):
        print('run_count : ', i)
        pathlist = make_path_list()
        for path in pathlist:
            candidate_dist = evaluate_path(path)
            if candidate_dist < dist:
                best_path = path
                dist = candidate_dist
    f = open(solution_file, 'w')
    for i in best_path:
        f.write(str(i) + '\n')
    f.close()
    return dist
'''
for path in path_list:
    dist = evaluate_path(path)
    if dist < best_dist:
        best_path = path
        best_dist = dist

    print("최적 거리: ", best_dist)
'''
```

```python
final_dist = find_best_path(GA_Interation)

f = open(solution_file, "r")
rdr = csv.reader(f)
final_path = []
for line in rdr:
  final_path.append(line)

f.close()
end_time = time.time()

print("최적 경로: ", final_path)
print("최적 거리: ", final_dist)
print("걸린 시간: " + f"{end_time-start_time}" + "sec")
```
